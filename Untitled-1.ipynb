```python
mapping = {'yes': 1, 'no': 0}

# Replace values in the DataFrame
# df['account status'] = df['account status'].replace(mapping)
df['upgrade status'] = df['upgrade status'].replace(mapping)
df["Borrower's inability"] = df["Borrower's inability"].replace(mapping)
df['Telephonic communication'] = df['Telephonic communication'].replace(mapping)
df['SARFAESI action required?'] = df['SARFAESI action required?'].replace(mapping)
df['Visit done?'] = df['Visit done?'].replace(mapping)
df['Recovered?'] = df['Recovered?'].replace(mapping)
df['Follow up done?'] = df['Follow up done?'].replace(mapping)
df['13(2) issued'] = df['13(2) issued'].replace(mapping)
df['Willingness to obtain OTS'] = df['Willingness to obtain OTS'].replace(mapping)
df['Advised to obtain insurance?'] = df['Advised to obtain insurance?'].replace(mapping)

```

```python
df.head()
```

```python
# Define the mapping
mapping = {'yes': 1, 'no': 0}

# Loop through each column and replace values
for column in ['upgrade status', "Borrower's inability", 
               'Telephonic communication', 'SARFAESI action required?', 'Visit done?', 
               'Recovered?', 'Follow up done?', '13(2) issued', 
               'Willingness to obtain OTS', 'Advised to obtain insurance?']:
    # Check unique values before replacement
    print(f"Unique values in {column} before replacement:", df[column].unique())
    
    # Replace values in the DataFrame
    df[column] = df[column].replace(mapping)
    
    # Check unique values after replacement
    print(f"Unique values in {column} after replacement:", df[column].unique())
```

```python
import matplotlib.pyplot as plt
status_counts = df['account status'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Account Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Group by 'account status' and 'SARFAESI action required?' and count occurrences
grouped_data = df.groupby(['account status', 'SARFAESI action required?']).size().unstack(fill_value=0)

# Plot the bar chart
grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add title and labels
plt.title('Bar Chart of Account Status vs. SARFAESI Action Required')
plt.xlabel('Account Status')
plt.ylabel('Count')

# Show the plot
plt.legend(title='SARFAESI Action Required')
plt.show()
```

```python
#VISUALIZATION OF ACTION REQUIRED IMMEDIATELY
import seaborn as sns
# Filter rows where both Borrower's inability and SARFAESI action required are 1
both_condition_rows = df[(df["Borrower's inability"] == 1) & (df["SARFAESI action required?"] == 1)]

# Plot the distribution of Efforts for rows where both conditions are satisfied
plt.figure(figsize=(12, 6))
sns.countplot(data=both_condition_rows, y='Efforts', order=both_condition_rows['Efforts'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('Efforts')
plt.title('Distribution of Efforts where Both Conditions are 1')
plt.show()
```

```python
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

    sarfaesi_time_series = df[df['SARFAESI action required?'] == 1].groupby(df['Date'].dt.to_period('M')).size()

    plt.figure(figsize=(12, 6))
    sarfaesi_time_series.plot(kind='line', marker='o', color='b')

    # Add title and labels
    plt.title('Time Series Analysis of SARFAESI Action Requirement')
    plt.xlabel('Date')
    plt.ylabel('Count of SARFAESI Action Required')
    plt.grid(True)

    # Show the plot
    plt.tight_layout()
    plt.show()
else:
    print("The 'Date' column is not available for time series analysis.")
```
